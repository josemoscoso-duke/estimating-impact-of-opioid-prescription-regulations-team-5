```python
import pandas as pd
import glob

path = r'E:/Jose_Moscoso/Duke/IDS 690 Tools for DS/Classes/Team Project/Population' # path of .csv files
all_files = glob.glob(path + "/*.csv")
Population = []

for filename in all_files:
    df_full = pd.read_csv(filename, sep = ';') 
    Population.append(df_full)

Population = pd.concat(Population, axis=0, ignore_index=True)
Population.describe()
Population.shape

Population.to_csv('E:/Jose_Moscoso/Duke/IDS 690 Tools for DS/Classes/Team Project/population_aggregate.csv')
```
